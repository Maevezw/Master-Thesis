```python
#importing packages
import pandas as pd
import json
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
#Loading all products dataframes and concatenate to two separate sets
def concat():
    for x in [1]:
        with open('albums-full-info-1.json') as json_file:
            data = json.load(json_file)
        d = json.loads(data)
        prods_x = pd.DataFrame.from_dict(d)
        print('finished set 1')
    for y in range(2,7):
        print('starting set', y)
        file_name = 'albums-full-info-'+str(y)+'.json'
        
        with open(file_name) as json_file:
            data = json.load(json_file)
        d = json.loads(data)
        prods_y = pd.DataFrame.from_dict(d)
        
        prods_x = pd.concat([prods_x, prods_y], ignore_index=True)
        print('finsished set', y)
    prods_x.to_csv('prods_1_of_2.csv')
    print('file 1 saved')
    return prods_x
```
