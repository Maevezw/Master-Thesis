```python
#import packages
import pandas as pd
import numpy as np 

pd.set_option('display.max_columns', None)
```

```python
#load data set
df = pd.read_csv('complete_set.csv')
```

```python
df.info()
```

```python
#remove all rows where products were not found in product data
df = df[df['_id_y'].notna()]
```

```python
df.info()
```

```python
#function to check how many unique values are found across all columns, and how many NaN values. 
def nunique(lst):
    for x in lst: 
        print('col_name:', x, ', number of unique values:' ,df[x].nunique(), ', number of nan:' ,df[x].isna().sum())
```

```python
cols = df.columns
nunique(cols)
```
